```python
import numpy as np
import pandas as pd
import re
import os
import warnings
import tensorflow as tf

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("./race_gender_data/df.csv").iloc[:, 1:]
df
```

```python
import joblib

# Load the pre-trained model
loaded_model = joblib.load("./gender/gender_predictor.pkl")


# Define a function to predict gender
def predict_gender(name):
    name = [name.lower()]  # Prepare the name for prediction
    predicted_gender = loaded_model.predict(name)  # Use the loaded model for prediction
    return (
        "Male" if predicted_gender[0] == 0 else "Female"
    )  # Return the predicted gender


df["predicted_gender"] = df["fullname"].apply(predict_gender)
df
```
